```python
# Import dependencies
import sqlite3
import pandas as pd
import numpy as np
from numpy import nan
```

```python
# Connect to database
conn = sqlite3.connect('database.sqlite')
```

# Query for Male Data

```python
# Read sqlite query results into a pandas DataFrame
male_df = pd.read_sql_query("""
SELECT 
*
FROM east_central_table

where gender = 'Male'

union

SELECT 
*
FROM north_east_table

where gender = 'Male'

union

SELECT 
*
FROM pacific_table

where gender = 'Male'

union

SELECT 
*
FROM south_table

where gender = 'Male'

union

SELECT 
*
FROM south_east_table

where gender = 'Male'

union

SELECT 
*
FROM south_west_table

where gender = 'Male'

union

SELECT 
*
FROM west_central_table

where gender = 'Male'

""",conn)

male_df.head()
```

# Query for Femal Data

```python
# Read sqlite query results into a pandas DataFrame
female_df = pd.read_sql_query("""
SELECT 
*
FROM east_central_table

where gender = 'Female'

union

SELECT 
*
FROM north_east_table

where gender = 'Female'

union

SELECT 
*
FROM pacific_table

where gender = 'Female'

union

SELECT 
*
FROM south_table

where gender = 'Female'

union

SELECT 
*
FROM south_east_table

where gender = 'Female'

union

SELECT 
*
FROM south_west_table

where gender = 'Female'

union

SELECT 
*
FROM west_central_table

where gender = 'Female'

""",conn)

female_df.head()
```

```python
# Save clean data to csv
male_df.to_csv('clean_data/male_data_2013_to_2017.csv', index = False)
female_df.to_csv('clean_data/female_data_2013_to_2017.csv', index = False)
```
